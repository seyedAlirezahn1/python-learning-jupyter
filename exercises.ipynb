Factorial Calculation

Write a program that takes a number n as input and calculates its factorial.

Factorial of n (written as n!) is the product of all numbers from 1 to n:

n! = 1 × 2 × 3 × ... × n

Input:
A single integer n on one line.
1 ≤ n ≤ 10

Output:
Print the value of n factorial.

Example:
Input: 5
Output: 120

````python
n = int(input())
ans = 1
for i in range(1,n+1):
  ans *= i

print(ans)
````

Right Triangle Check

Write a program that takes three positive integers as input and prints YES if they can form a right triangle, otherwise prints NO.

In a right triangle, the square of the longest side equals the sum of the squares of the other two sides. (Pythagorean theorem: a² + b² = c²)

Input:
Three positive integers a, b, c, each on a separate line.
1 ≤ a, b, c ≤ 150

Output:
Print YES if the three numbers can form a right triangle, otherwise NO.

Example 1:
Input:
3
4
5
Output: YES

Example 2:
Input:
2
3
4
Output: NO

````python
a = int(input())
b = int(input())
c = int(input())

if a>b and a>c:
  if (b**2)+(c**2)==a**2:
    print("yes")
  else:
    print("no")
elif b>c and b>a:
  if (a**2)+(c**2)==b**2:
    print("yes")
  else:
    print("no")
else:
  if (a**2)+(b**2)==c**2:
    print("yes")
  else:
    print("no")
````

Digit Printer

Write a program that takes an integer with an unknown number of digits and prints each digit repeated as many times as its value.

Input:
A single integer on one line. The length of the number is less than 100.

Output:
For each digit, print the digit itself followed by `: ` and then repeat that digit as many times as its value.

Example:
Input: 50943
Output:
5: 55555
0:
9: 999999999
4: 4444
3: 333

````python
n = input()

for digit in n:  
    count = int(digit)  
    print(digit, ":", end=" ")
    for _ in range(count):
        print(digit, end="")
    print()  
````

Single Digit Reducer

Mehdi is tired of coding and no longer has the patience for numbers with more than one digit. So he converts any multi-digit number into a single digit using his own method.

He replaces the number with the sum of its digits to get a new number. Then he does the same with the new number and keeps repeating until he reaches a single digit. After a while, Mehdi realized this didn't make things easier — he was even more caught up with numbers. So he asked you to help him reduce numbers to a single digit.

Input:
A single integer n on one line.
1 ≤ n ≤ 10^18

Output:
Print the single digit obtained by repeatedly summing the digits of n until only one digit remains.

Example 1:
Input: 14
Output: 5

Example 2:
Input: 123456
Output: 3
Step 1: 1+2+3+4+5+6 = 21
Step 2: 2+1 = 3

````python
n = int(input())

while n >= 10: 
    total = 0
    temp = n
    while temp > 0:
        total += temp % 10 
        temp //= 10 
    n = total 

print(n)
````

Perfect Number Check

Write a program that takes a number n as input and prints YES if it is a perfect number, otherwise prints NO.

A positive integer n is called perfect if and only if the sum of its proper divisors (divisors smaller than n) equals n.

Input:
A single integer n on one line.
2 ≤ n ≤ 200,000

Output:
Print YES if the number is perfect, otherwise print NO.

Example 1:
Input: 27
Output: NO
Proper divisors of 27: 1, 3, 9
Sum: 1 + 3 + 9 = 13 ≠ 27

Example 2:
Input: 6
Output: YES
Proper divisors of 6: 1, 2, 3
Sum: 1 + 2 + 3 = 6

````python
a = int(input())
result = 0

for i in range(1,a):
  if a%i==0:
    result += i

if result == a:
  print("yes")
else:
  print("no")
````

Next Power of Two

Write a program that takes a number n as input and prints the first power of 2 that is strictly greater than n.

Input:
A single integer n on one line.
1 ≤ n ≤ 10^9

Output:
Print the smallest power of 2 that is greater than n.

Example 1:
Input: 95
Output: 128
2^6 = 64 ≤ 95 < 128 = 2^7

Example 2:
Input: 1024
Output: 2048
2^10 = 1024 ≤ 1024 < 2048 = 2^11

````python
n = int(input())
result = 2

while result <= n:
  result *= 2

print(result)
````

Reverse Until Zero

Write a program that reads numbers from input repeatedly until the digit zero is entered. After zero is entered, the program should print all the numbers that were entered before zero, in reverse order.

Input:
Each line contains a single integer. The number of inputs is less than 1000. It is guaranteed that zero appears only once at the end of the input. All other numbers are natural numbers up to 1000.

Output:
Print each number on a separate line in reverse order of input.

Example:
Input:
3
4
7
4
9
0
Output:
9
4
7
4
3

````python
a = None
ans = []

while a != 0:
  a = int(input())
  ans.append(a)

ans.remove(0)

for i in range(len(ans)-1,-1,-1):
  print(ans[i])
````

The Donkey's Concert

One day, a donkey from the desert went to a football stadium and, due to his donkey-ness, ended up on the field instead of the stands. Seeing the grass, he got excited, and seeing thousands of people around him doubled his excitement. He felt like he was at a concert and decided to sing (bray) for them! He knew nothing about music or rhythm, but to make his braying rhythmic, he decided to put specific intervals between his brays.

He chose two numbers, a and b, and decided to perform as follows:

At second 0, he announces that he will sing. Then he waits a seconds and brays for the first time at second a. Then he waits b seconds and brays for the second time at second a+b. Then he waits a seconds again and brays at second 2a+b. Then b seconds, and so on...

He had decided beforehand not to bray more than l times (his throat can't handle more). Now he has brayed l times and is wondering how many seconds have passed since he made the announcement until his last bray. He is a donkey and asks you to help him figure it out.

Input:
A single line containing three integers a, b, and l — the waiting times and the number of brays.
1 ≤ a, b, l ≤ 1000

Output:
Print the time of the last bray.

Example 1:
Input: 1 1 1
Output: 1

Example 2:
Input: 3 4 5
Output: 17

Example 3:
Input: 10 3 2
Output: 13

````python
a, b, l = map(int, input().split())
total = (l // 2) * (a + b) + (l % 2) * a
print(total)
````

Chess Pieces

Pouyan is a chubby teenager who has decided to lose weight by playing chess.

He borrowed a chessboard and pieces from the chess masters to exercise. But he realized that the number of pieces he was given is incorrect — some pieces are too many, and some are too few.

We know that a proper chess set must include:

- One King
- One Queen
- Two Rooks
- Two Bishops
- Two Knights
- Eight Pawns

Pouyan can add or remove pieces to make the set correct. Given the number of each piece he currently has, determine how many of each he needs to add or remove.

Input:
A single line with 6 integers representing the number of Kings, Queens, Rooks, Bishops, Knights, and Pawns he currently has.

Output:
A single line with 6 integers. For each piece type, print a positive number if he needs to add that many, or a negative number if he needs to remove that many.

Example:
Input: 0 2 2 2 3 7
Output: 1 -1 0 0 -1 1

Explanation:
King: 0 → needs 1 (add 1)
