Factorial Calculation

Write a program that takes a number n as input and calculates its factorial.

Factorial of n (written as n!) is the product of all numbers from 1 to n:

n! = 1 × 2 × 3 × ... × n

Input:
A single integer n on one line.
1 ≤ n ≤ 10

Output:
Print the value of n factorial.

Example:
Input: 5
Output: 120

```python
n = int(input())
ans = 1
for i in range(1,n+1):
  ans *= i

print(ans)
```

Right Triangle Check

Write a program that takes three positive integers as input and prints YES if they can form a right triangle, otherwise prints NO.

In a right triangle, the square of the longest side equals the sum of the squares of the other two sides. (Pythagorean theorem: a² + b² = c²)

Input:
Three positive integers a, b, c, each on a separate line.
1 ≤ a, b, c ≤ 150

Output:
Print YES if the three numbers can form a right triangle, otherwise NO.

Example 1:
Input:
3
4
5
Output: YES

Example 2:
Input:
2
3
4
Output: NO

```python
a = int(input())
b = int(input())
c = int(input())

if a>b and a>c:
  if (b**2)+(c**2)==a**2:
    print("yes")
  else:
    print("no")
elif b>c and b>a:
  if (a**2)+(c**2)==b**2:
    print("yes")
  else:
    print("no")
else:
  if (a**2)+(b**2)==c**2:
    print("yes")
  else:
    print("no")
```

Triangle Angles Check

Baqer has a cold and is very tired.

Baqer has always loved geometry since he was a kid. Tonight's homework is to take 3 angles and determine whether a triangle can be formed with them.

You are given the three angles that the teacher gave to Baqer. Help him solve his homework correctly.

Input:
Three integers on the first line representing the degrees of the 3 angles.
All three numbers are guaranteed to be non-negative integers less than 360.

Output:
Print Yes if a triangle can be formed with these 3 angles, otherwise print No.

Hint: A triangle is valid if all three angles are greater than 0 and their sum equals 180.

Example 1:
Input: 60 70 50
Output: Yes

Example 2:
Input: 90 90 0
Output: No

```python
a = input()

l = a.split(" ")

if int(l[0]) + int(l[1]) +int(l[2]) == 180 and int(l[0]) != 0 and int(l[1]) != 0 and int(l[2]) != 0:
  print("yes")
else:
  print("no")
```

Multiplication Table

Write a program that takes a number n as input and prints the multiplication table from 1 to n.

Input:
A single integer n on one line.
1 ≤ n ≤ 100

Output:
Print n lines, each containing the multiplication table for that row up to n.

Example:
Input: 5
Output:
1 2 3 4 5
2 4 6 8 10
3 6 9 12 15
4 8 12 16 20
5 10 15 20 25

```python
a = int(input())
for i in range(1,a+1):
  for j in range(1,a+1):
    print(j*i,end=" ")
  print(" ")
```
